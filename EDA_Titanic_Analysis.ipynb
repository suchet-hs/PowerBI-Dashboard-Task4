```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
df.describe()
```

Observation:
* Dataset contains missing values in Age, Cabin, Embarked
* Numerical features include Age, Fare, etc.

```python
df['survived'].value_counts()
df['sex'].value_counts()
df['pclass'].value_counts()
```

Observation:
* More males than females
* Majority passengers are in 3rd class

```python
df.isnull().sum()
```

Observation:
* Age and Cabin have many missing values

```python
df['age'].hist(bins=30)
plt.title("Age Distribution")
plt.show()
```

Observation:
* Most passengers are between 20–40 years

```python
sns.boxplot(x='pclass', y='fare', data=df)
plt.show()
```

```python
Observation:
* 1st class passengers paid higher fares
```

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.show()
```

Observation:
* Females survived more than males

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

Observation:
* Fare and Pclass are correlated

```python
sns.pairplot(df[['age','fare','pclass','survived']].dropna())
plt.show()
```

Observation:
* Survival depends on class and fare

Summary :
* Females had higher survival rate
* Higher class passengers survived more
* Age group 20–40 dominates dataset
* Fare and class influence survival
* Missing values present in Age & Cabin
